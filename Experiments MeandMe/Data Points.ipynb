```python
path = "./data/dichalcogenides_public"
struc = path + "/structures/"
```

```python
import pandas as pd
import json
from pymatgen.core import Structure
import numpy as np
```

```python
df = pd.read_csv(path+"/targets.csv")
df.head()
```

```python
for ind, row in df.iterrows():
    print(row["_id"])
    getpath = struc + row["_id"]+ ".json"
    with open(getpath) as f:
        json_df = json.load(f)
    print(getpath)
    break
```

```python
struct_s = Structure.from_dict(json_df)
# struct_s
```

```python
def break_sites(df):
    arr = []
    for ind, row in df.iterrows():
        for elem in row.tolist():
            arr.append(elem)
    return arr
# break_sites(struct_s.as_dataframe())
```

```python
mat = []
for ind in df.index:
    arr = []
#     print(df["_id"][ind])
    getpath = struc+df["_id"][ind] +".json"
    with open(getpath) as f:
        json_ = json.load(f)
        json_struct = Structure.from_dict(json_)
#     print(getpath)
    
    
    
    arr.append("target:")
    arr.append(df["band_gap"][ind])
    
    
    arr.append("charge:")
    arr.append(json_struct.charge)
    
    
    
    arr.append("lattice:")
    latt_lis = json_struct.lattice.as_dict()["matrix"]
    for elem in latt_lis:
        arr.extend(elem)
        
        
        
        
    arr.append("sites:")
    site_df = json_struct.as_dataframe()
#     print(type(site_df))
    arr.extend(break_sites(json_struct.as_dataframe()))
#     print(arr)

    
    
    mat.append(arr)
```

```python
final_df = pd.DataFrame(mat)
```

```python
len_arr = []
for row in mat:
    len_arr.append(len(row))
# len_arr
```

```python
final_df["length"] = len_arr
```

```python
final_df.to_csv("data_mod1.csv")
```
